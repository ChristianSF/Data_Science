```python
import pandas as pd
import numpy as np
from scipy.spatial import ConvexHull, convex_hull_plot_2d
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv("/content/postos_itap - postos_itap.csv")
dados.shape
```

```python
dados.columns
```

```python
coords = ['LATITUDE', 'LONGITUDE']
postos = dados[coords]
```

```python
postos = np.asarray(postos)
```

```python
hull = ConvexHull(postos)
```

```python
plt.plot(postos[:,0], postos[:,1], 'o')
for simplex in hull.simplices:
    plt.plot(postos[simplex, 0], postos[simplex, 1], 'k-')
```

```python
plt.plot(postos[hull.vertices,0], postos[hull.vertices,1], 'r--', lw=2)
plt.plot(postos[hull.vertices[0],0], postos[hull.vertices[0],1], 'ro')
plt.show()
```
