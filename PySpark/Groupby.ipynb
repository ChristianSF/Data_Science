```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('Group').getOrCreate()
```

```
dados = spark.read.csv("/FileStore/tables/sales_info.csv", header=True, inferSchema=True)
```

```
dados.show()
```

```
dados.printSchema()
```

```
dados.groupBy("Company")
```

```
dados.groupBy("Company").mean().show()
```

```
dados.groupBy("Company").count().show()
```

```
dados.agg({'Sales':'sum'}).show()
```

```
dados.agg({'Sales':'max'}).show()
```

```
group_data = dados.groupBy("Company")
```

```
group_data.agg({'Sales':'max'}).show()
```

```
from pyspark.sql.functions import countDistinct, avg, stddev
```

```
dados.select(countDistinct('Sales')).show()
```

```
dados.select(avg('Sales')).show()
```

```
dados.select(avg('Sales').alias('Average Sales')).show()
```

```
dados.select(stddev('Sales')).show()
```

```
from pyspark.sql.functions import format_number
```

```
sales_std = dados.select(stddev("Sales").alias('std'))
```

```
sales_std.show()
```

```
sales_std.select(format_number('std', 3).alias('std')).show()
```

```
dados.show()
```

```
#Ordenar de forma crescente
dados.orderBy("Sales").show()
```

```
#Ordenar de forma descrescente
dados.orderBy(dados['Sales'].desc()).show()
```
