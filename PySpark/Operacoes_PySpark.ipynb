```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName("Ops").getOrCreate()
```

```
dados = spark.read.csv("/FileStore/tables/appl_stock.csv", inferSchema=True, header=True)
```

```
dados.show()
```

```
dados.printSchema()
```

```
dados.head(1)[0]
```

```
dados.filter("Close < 500").show()
```

```
dados.filter("Close < 500").select(['Open', 'Close']).show()
```

```
dados.filter(dados['Close'] < 500).show()
```

```
dados.filter(dados['Close'] < 500).select('Volume').show()
```

```
dados.filter((dados['Close'] < 200) & (dados['Open']>200)).show()
```

```
dados.filter((dados['Close'] < 200) &~(dados['Open']>200)).show()
```

```
dados.filter(dados['Low'] == 197.16).show()
```

```
dados.filter(dados['Low'] == 197.16).collect()
```

```
resultado = dados.filter(dados['Low'] == 197.16).collect()
row = resultado[0]
```
